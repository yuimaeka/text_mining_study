```python
from janome.tokenizer import Tokenizer     # for janome tokenizer
from janome.analyzer import Analyzer       # for janome anolyzer
from janome.charfilter import *            # for janome character filter
from janome.tokenfilter import *           # for janome token filter
from collections import Counter  # for Counter
import codecs                    # for file I/O
import glob                      # for file handling
from sklearn.feature_extraction.text import CountVectorizer as CV    # for CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer as TV    # for TfidfVectorizer
import re      # for regular expression
```

```python
file_list = glob.glob("*")
file_list
```

```python
documents = []
for file in file_list:
    with codecs.open(file, 'r',  'utf8') as f:
        documents.append(f.read())
        
print(len(documents))
```

my_analyzer()を定義する

```python
stop_words =['rt']

def isStopWords(token, stop_list):
    if (token in stop_list) or (re.match(r'(^[a-zA-Z0-9_ぁ-んぁ-ン]$)|(^ー+\w*$)|(^[-$!#%"\'()\[\]{}\\|~@^=+*/;:<>,.&]$)', token)):
        return True
    else:
        return False


def my_analyzer(string):
    char_filters = [UnicodeNormalizeCharFilter(), RegexReplaceCharFilter('(http(s)?://([-\w]+\.)+[-\w]+(/[-\w ./?%&=]*)?)|(@([\w]+))|(!+|(ww+))', '')]
    token_filters = [CompoundNounFilter(), POSStopFilter(['名詞,代名詞','名詞,数','記号']), LowerCaseFilter(), POSKeepFilter(['名詞,一般']), ExtractAttributeFilter(att = 'base_form')]
    a = Analyzer(char_filters = char_filters, token_filters=token_filters)
    tokens = list(a.analyze(string))
    #return [token for token in tokens if (token not in stop_words) and (not re.match(r'(^[a-zA-Z0-9_ぁ-んぁ-ン]$)|(^ー+\w*$)|(^[-@$!#%"\'()\[\]{}\\|~^=+\*/;:<>,.&])', token))]
    return [token for token in tokens if not isStopWords(token, stop_words)]
```

CountVectrizer と TfidfVectorizer による単語文書行列の生成

```python
cv = CV(analyzer=my_analyzer, min_df=0.1, max_df=0.8)
matrix = cv.fit_transform(documents)
matrix
```

```python
tv = TV(analyzer=my_analyzer, min_df=0.1, max_df=0.8)
matrix_tfidf = tv.fit_transform(documents)
matrix_tfidf
```

vocabulary の作成（1）
各単語の各文書での出現頻度・Tf-IDF値のうち最大のものを抽出し，その上位100個を利用

```python
# 圧縮形式の行列表現を通常の行列表現に変換
res = matrix.toarray()

# 単語の取得
features = cv.get_feature_names()

# 各単語の出現頻度の最大値をリスト化
max_cv_list = []
for i in range(len(cv.get_feature_names())):
    w_list = []
    for j in range(40):
        w_list.append(res[j][i])
    max_cv_list.append(max(w_list))

# 各単語を出現頻度の最大値で降順にソート
zipped_list = [(v, i) for i, v in zip(features, max_cv_list)]
desc_list = sorted(zipped_list, reverse=True)

# 出現頻度の最大値に関して上位100件を vocabulary として登録
cv_vocab = []
for j in range(100):
    print('  ' + str(desc_list[j]))
    cv_vocab.append(desc_list[j][1])
```

これ（全体）だと地震関係の単語が多くなってしまう。偏りがでる。
なので、各文書について上位五個を取ってくるやり方に変えたい！

vocabulary の制限方法を各文書における頻度上位10単語をマージしたものにしてみる

```python
res = matrix.toarray()
features = cv.get_feature_names()
cv_vocab2 = []
for i in range(40):
    print(file_list[i])
    ######### for the results of  CV ##########
    print('### Word Count ###')
    zipped_list = [(v, j) for j, v in zip(features, res[i])]
    desc_list = sorted(zipped_list, reverse=True)
    for j in range(10):
        print('  ' + str(desc_list[j]))
        cv_vocab2.append(desc_list[j][1])
```

```python
# 生成した vocabulary の確認
cv_vocab2
```

```python
# この時点では各文書から10個ずつ抽出しただけなので，重複も含めて400個の単語があることを確認
len(cv_vocab2)
```

```python
# set() で一旦集合に変換することで重複要素を削除し，さらに list() でリストに戻す
cv_vocab2 = list(set(cv_vocab2))
# 上記の操作で重複要素が削除され，最終的に204個の単語のリストとなっていることを確認
len(cv_vocab2)
```

vocabulary の制限方法を各文書における Tf-Idf スコア上位10単語をマージしたものにしてみる

```python
res_tfidf = matrix_tfidf.toarray()
features_tfidf = tv.get_feature_names()
tv_vocab2 = []
for i in range(40):
    print(file_list[i])
    ######### for the results of  TV ##########
    print('### Tfidf weight ###')
    zipped_list = [(v, i) for i, v in zip(features_tfidf, res_tfidf[i])]
    desc_list = sorted(zipped_list, reverse=True)
    for j in range(10):
        print('  ' + str(desc_list[j]))
        tv_vocab2.append(desc_list[j][1])

print(len(tv_vocab2))
tv_vocab2 = list(set(tv_vocab2))
# 重複する単語を削除した結果，287個の単語のリストになっていることを確認
print(len(tv_vocab2))
```

可視化したい。そのために二次元化したい。
主成分分析する。

主成分分析（PCA: Principal Component Analysis) による次元圧縮

n次元ベクトルを m (<= n) 次元ベクトルに変換する技術

n次元ベクトルの集合（データセット）から主成分と呼ばれるベクトルを計算

第1主成分～第n主成分まで存在

各主成分はn次元ベクトル

元のベクトルは各主成分の要素を係数とした線形和を要素とする m 次元ベクトルに変換

変換後のベクトルの1次元目の要素には第1主成分を利用

2次元目の要素には第2主成分を利用

例）

第1主成分 (p_11, p_12, ..., p_1n)

第2主成分 (p_21, p_22, ..., p_2n)


元のベクトル (v_1, v_2, ...., v_n)


元のベクトルの2次元ベクトルへの変換

(p_11 × v_1 + p_12 × v_2 + ... + p_1n × v_n, p_21 × v_1 + p_22 × v_2 + ... + p_2n × v_n)

```python
# PCA 用のライブラリのインポート
from sklearn.decomposition import PCA

# 2次元ベクトルに削減するインスタンスの生成
pca = PCA(n_components=2)
```

```python
res=matrix_cv2.toarray()
# 行列 res の各行ベクトルを2次元に圧縮するための主成分の計算
pca.fit(res)
```

```python
# 元の行列を実際に2次元（40行2列）に圧縮した結果を求める
res_pca = pca.fit_transform(res)
```

```python
# 数値データ計算ライブラリである numpy の読み込み．以降，np として利用
import numpy as np
# データのグラフを描画するライブラリである matplotlib から pyplot を読み込み．以降，plt として利用．
from matplotlib import pyplot as plt
```

転置行列にしなきゃいけない。

```python
# 2次元に圧縮後の行列の転置行列を .T で求め，2行40列の行列に変換したのち，その各行を Xベクトル，Yベクトルとする．
# 各ベクトルが，各データの2次元座標の x 座標， y 座標となる．
X = res_pca.T[0]
Y = res_pca.T[1]
```

```python
# Xベクトルの確認
X
```

```python
# Yベクトルの確認
Y
```

```python
# 2次元に圧縮後のデータを散布図として2次元平面に描画
# plt.scatter() で散布図の生成
# plt.show() で実際に画面上に描画
plt.scatter(X, Y)
plt.show()
```

```python
# クラスタリング結果のラベルが .labels_ で参照可能であることの確認
cls.labels_
```

```python
# 各クラスごとにデータ点の色を変えて描画
# enumerate() は引数で与えられたリストのインデックスと対応する値のペアを返す関数
# ここでは，i がインデックス，label が対応する値
# plt.scatter() の c オプションは色（color）の指定
# 以下の for文では，データ点を1つずつ描画している
# plt.show() を実行するまでは，plt.scatter() は何度実行してもいい．同一平面上に重ねて描画される．
for (i, label) in enumerate(cls.labels_):
    if label == 0:
        plt.scatter(X[i], Y[i], c='red')
    elif label == 1:
        plt.scatter(X[i], Y[i], c='blue')
    elif label == 2:
        plt.scatter(X[i], Y[i], c='green')
plt.show()
```
