```python
import io
labels = []
for i in range(5):
    fname = 'tag' + str(i+1) + '.txt'
    print(fname)
    with io.open(fname, 'r', encoding = 'utf_8_sig') as f:
        line = f.readline()
        while line:
            labels.append(line.strip())
            line = f.readline()
len(labels)   
```

##### ラベルの頻度集計

```python
from collections import Counter
dist = Counter(labels)
for label in dist:
    print(label)
```

```python
dist.most_common
```

数個しかないラベルの記事はあまり特徴がないのでその他にまとめる。
訓練に出てテストに出ないことも！今回は五個未満


```python
# 「その他」に変更するラベルをリストにまとめる
replaced = ['重量挙げ', 'スノーボード', 'スピードスケート', '卓球', 'カーリング', '高校野球', 'ラグビー', 'アメリカンフットボール', 'ボート', 'テコンドー', 'アイスホッケー', '馬術', '少林寺拳法', '新体操', 'アーチェリー', 'フェンシング']

# enumerate() はリストのインデックスと対応する値のペアを返す
for i, v in enumerate(labels):
    # ラベルの値が replaced の中のいずれかと一致する場合，「その他」に置換
    if v in replaced:
        print(str(i) + ': ' + labels[i])
        labels[i] = 'その他'
        print(str(i) + ': ' + labels[i])
```

```python
dist = Counter(labels)
dist.most_common
```

##### ラベルのリスト中の文字列を数値に置き換える

手順1：Counter() の結果からラベルを取り出し，リストにまとめる => label_list

```python
label_list = []
for v in dist:
    label_list.append(v)
```

```python
label_list
```

```python
print(label_list.index('相撲'))
```

手順2：labels 中の各ラベルを label_list における対応するインデックスの値に置換


index() メソッドは引数に指定した要素のインデックスを返す（リスト中の要素に重複がある場合は注意する）

```python
for i, v in enumerate(labels):
    labels[i] = label_list.index(v)

# 置換後の数値ラベルの頻度分布を確認しておく
dist_id = Counter(labels)
dist_id.most_common
```

手順1. ファイルの読み込み
以下のファイル読み込みは，読み込み順序が正しくない．

ファイル名の生成
ソートで解決するのは難しいのでファイル名中の番号部分を range() で生成してリスト化する

```python
new_file_list = []
prefix = 'article_'
extension = '.txt'

# ファイル名リストの生成
for i in range(900):
    new_file_list.append(prefix + str(i+1) + extension)
new_file_list
```

```python
documents = []
for file in new_file_list:
    with codecs.open(file, 'r',  'utf8') as f:
        documents.append(f.read())
        
print(len(documents))
```

my_analyzer() の設定
（注意事項）POSKeepFilter() で残す品詞を指定する場合，一般名詞に限定しない．一般名詞に限定すると，個人名などが排除され，分類がうまくいかなくなる．

isStopWords() 中の正規表現は数値を含む単語，半角記号だけを含む単語を省くように修正（ただし，一部うまくいっていない）

```python
stop_words =['rt', 'sports', 'watch']

def isStopWords(token, stop_list):
    if (token in stop_list) or (re.match(r'(^[0-9]+\w*$)|(^[a-zA-Z0-9_ぁ-んぁ-ン]$)|(^ー+\w*$)|(^\w*[-$!#%"\'()\[\]{}\\|~@^=+*/;:<>,.&]+\w*$)', token)):
        return True
    else:
        return False


def my_analyzer(string):
    char_filters = [UnicodeNormalizeCharFilter(), RegexReplaceCharFilter('(http(s)?://([-\w]+\.)+[-\w]+(/[-\w ./?%&=]*)?)|(@([\w]+))|(!+|(ww+))', '')]
    token_filters = [CompoundNounFilter(), POSStopFilter(['名詞,代名詞','名詞,数','記号']), LowerCaseFilter(), POSKeepFilter(['名詞']), ExtractAttributeFilter(att = 'base_form')]
    a = Analyzer(char_filters = char_filters, token_filters=token_filters)
    tokens = list(a.analyze(string))
    #return [token for token in tokens if (token not in stop_words) and (not re.match(r'(^[a-zA-Z0-9_ぁ-んぁ-ン]$)|(^ー+\w*$)|(^[-@$!#%"\'()\[\]{}\\|~^=+\*/;:<>,.&])', token))]
    return [token for token in tokens if not isStopWords(token, stop_words)]
```

```python
tv = TV(analyzer=my_analyzer, min_df=0.005, max_df=0.6)
matrix_tfidf = tv.fit_transform(documents)
matrix_tfidf
```

```python
print(tv.get_feature_names())
```

```python
res_tfidf = matrix_tfidf.toarray()
features_tfidf = tv.get_feature_names()
tv_vocab2 = []
for i in range(900):
    print(new_file_list[i])
    ######### for the results of  TV ##########
    print('### Tfidf weight ###')
    zipped_list = [(v, i) for i, v in zip(features_tfidf, res_tfidf[i])]
    desc_list = sorted(zipped_list, reverse=True)
    for j in range(5):
        print('  ' + str(desc_list[j]))
        tv_vocab2.append(desc_list[j][1])

print(len(tv_vocab2))
tv_vocab2 = list(set(tv_vocab2))
# 重複する単語を削除した結果，延べ数より単語数が減少していることを確認
print(len(tv_vocab2))
# このあとは，CV で tv_vocab2 をつかって行列を作り直す
```

```python
# ライブラリの読み込み
from sklearn.metrics import confusion_matrix
```

```python
# 予測結果の保存
result = clf.predict(matrix_cv)
result
```

```python
# 混同行列（confusion matrix）の生成
confusion_matrix(labels, result)
```

```python
# 混同行列の保存
conf_matrix = confusion_matrix(labels, result)

# 混同行列の行ごとに確認
#   クラス i に対する正解数：conf_matrix[i][i]
# 　クラス i に属する事例数：sum(conf_matrix[i])
for i in range(len(label_list)):
    # クラス i に対する正解率の計算
    print(label_list[i] + ': ' + str(conf_matrix[i][i]) + '/' + str(sum(conf_matrix[i])))
    
    # 間違い先のクラス j を確認して，その件数を表示
    for j in range(len(label_list)):
        if j != i and conf_matrix[i][j] > 0:
            print("  " + label_list[j] + ': ' + str(conf_matrix[i][j]))

# 全体の正解率（accuracy）の計算
# 　全体の正解数＝対角成分の総和：sum(conf_matrix.diagonal())
# 　事例の総数：len(labels)
# 　round()関数は値を丸めるため
print('accuracy = ' + str(round(sum(conf_matrix.diagonal()) / len(labels), 4)))
```
