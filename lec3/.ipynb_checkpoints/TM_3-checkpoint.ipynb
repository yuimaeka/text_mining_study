```python
import glob
file_list = glob.glob("*")
file_list
```

```python
documents=[]
for file in file_list:
    with codecs.open(file,'r','utf8') as f:
        documents.append(f.read())
        
print(len(documents))
```

```python
print(documents[0])
```

前回のアナライザーを使用。!（びっくり）やww（草）の表現も省くよう改良

```python
char_filters = [UnicodeNormalizeCharFilter(), RegexReplaceCharFilter('(http(s)?://([-\w]+\.)+[-\w]+(/[-\w ./?%&=]*)?)|(@([\w]+))|(!+|(ww+))', '')]
token_filters = [CompoundNounFilter(), POSStopFilter(['名詞,代名詞','名詞,数','記号']), LowerCaseFilter(), POSKeepFilter(['名詞,一般']), ExtractAttributeFilter(att = 'base_form')]
a = Analyzer(char_filters = char_filters, token_filters=token_filters)
```

```python
tc = a.analyze(documents[0])
for token in tc:
    print(token)
```

[pythonの正規表現操作](https://docs.python.jp/3/library/re.html)

```python
import re #正規表現を扱う為のモジュール

stop_words=['rt']

def isStopWords(token, stop_list):
    if(token in stop_list)or(re.match(r'(\b[a-zA-Z0-9_あ-んア-ンぁ-ぉァ-ォ]\b)|(\bー+\w*\b)',token)):
        return True
    else:
        return False
    


def my_analyzer(string):
    char_filters = [UnicodeNormalizeCharFilter(), RegexReplaceCharFilter('(http(s)?://([-\w]+\.)+[-\w]+(/[-\w ./?%&=]*)?)|(@([\w]+))|(!+|(ww+))', '')]
    token_filters = [CompoundNounFilter(), POSStopFilter(['名詞,代名詞','名詞,数','記号']), LowerCaseFilter(), POSKeepFilter(['名詞,一般']), ExtractAttributeFilter(att = 'base_form')]
    a = Analyzer(char_filters = char_filters, token_filters=token_filters)
    tokens = list(a.analyze(string))
    return [token for token in tokens if not isStopWords(token,stop_words)]
```

```python
from sklearn.feature_extraction.text import CountVectorizer as CV    # for CountVectorizer
cv = CV(analyzer=my_analyzer,min_df=0.4)
matrix = cv.fit_transform(documents[0:3])
matrix
```

```python
cv.get_feature_names()
```

min_df 最小限頻度いくつのドキュメントに出てくるか
sparse matrixはスカスカの行列

知ってる行列表現に直す

```python
res=matrix.toarray()
res[0]
```

```python
features = cv.get_feature_names()
for i in range(3):
    print('################ '+file_list[i]+' #################')
    for fe,count in zip(features,res[i]):
        print(fe+': '+str(count))
```

頻度順に並べ替える

```python
for i in range(3):
    print('################ '+file_list[i]+' #################')
    zipped_list=[(v,i) for i,v in zip(features,res[i])]
    desc_list=sorted(zipped_list, reverse=True)
    for j in range(10):
        print('  '+str(desc_list[j]))
```

40個のファイルで五個ずつ表示
max_df 最大頻度
全部のファイルに入っているようなのもいらない！

```python
cv = CV(analyzer=my_analyzer,min_df=0.1,max_df=0.8)
matrix = cv.fit_transform(documents)
res=matrix.toarray()
features = cv.get_feature_names()
for i in range(40):
    print('################ '+file_list[i]+' #################')
    zipped_list=[(v,i) for i,v in zip(features,res[i])]
    desc_list=sorted(zipped_list, reverse=True)
    for j in range(5):
        print('  '+str(desc_list[j]))
```

[tfidf](https://ja.wikipedia.org/wiki/Tf-idf) を使用するのもある
[tfidfvectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html)

一つの文章にしか出てきていないものは、特徴量が多いとして重要視する。

```python
from sklearn.feature_extraction.text import TfidfVectorizer as TV    # for TfidfVectorizer
tv = TV(analyzer=my_analyzer, min_df=0.1, max_df=0.8)
matrix_tfidf = tv.fit_transform(documents)
res_tfidf = matrix_tfidf.toarray()
features_tfidf = tv.get_feature_names()
for i in range(40):
    print('#####'+file_list[i]+'#####')
    zipped_list = [(v, i) for i, v in zip(features_tfidf, res_tfidf[i])]
    desc_list = sorted(zipped_list, reverse=True)
    for j in range(5):
        print('  ' + str(desc_list[j]))
```
